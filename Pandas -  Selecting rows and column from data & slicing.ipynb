# Pandas -  Selecting rows and column from data & slicing

```python
import pandas as pd
```

```python
df = pd.read_csv('/Users/Akshay/Documents/titanic_train.csv')
```

```python
df.head()
```

```python
# selecting column
df['Name'].head(10)
```

```python
#selecting multiple column, pass an array of column names
df[['Name','Fare']].head(10)
```

```python
# selecting a row using iloc and giving the index number of the row to be selected
df.iloc[1]
```

```python
# selecting multiple rows using iloc
df.iloc[4:9]
```

```python
# Selecting a perticular row and a column
df.iloc[4,3]
```

```python
# Selecting a perticular row and multiple column
df.iloc[4,3:6]
```

```python
df.iloc[4:11,3:6]
```
