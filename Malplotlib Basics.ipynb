# Bar chart

```python
x =[1,2,3,4,5]
y =[10,2,7,4,1]
plt.bar(x,y,width = .2)
plt.xlabel('x data')
plt.ylabel('y data')
plt.title('Simple bar plot')
plt.show()
```

# Histogram 
plot the distribution of data under data_bins (data intervals)

```python
import numpy as np
x = 40 + np.floor(60* np.random.rand(100))
x_bin = [40,50,60,70,80,90,100]
plt.hist(x,x_bin)
plt.xlabel('data bins/interval')
plt.ylabel('data counts')
plt.title('Simple histogram plot')
plt.show
```

# Scatter plot 

helpful for classification to see the how the data is distributed in classes or for regression to visualise how your data is distributed

```python
x = np.floor(100 * np.random.rand(100))
y = np.floor(30 * np.random.rand(100))
plt.scatter(x,y)
plt.xlabel('x data')
plt.ylabel('y data')
plt.title('Simple scatter plot')
plt.show()
```

# Stack plot
used when for each instance of x there can be multiple y values. Eg stock price on a day(x) for apple,google,facebook,microsoft


```python
days= [1,2,3,4,5]
apple = [234,543,212,343,123]
google = [432,345,352,213,531]
facebook = [100,90,96,94,134]
microsoft = [87,83,93,103,164]
```

```python
plt.stackplot(days,apple,google,facebook,microsoft)

plt.xlabel('day')
plt.title('Simple stackplot')
plt.show()
```
